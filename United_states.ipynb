```python
# import libraries

import requests
from bs4 import BeautifulSoup as bs
```

```python
# Sending http request to get the html content

link = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
page_response = requests.get(link)
print(page_response)

print("Successful")
```

```python
# Parsing the html content for data extraction

soup2 = bs(page_response.content,"html.parser")
print(soup2.prettify())
```

```python
# Analyze the html content and extract the useful data

table = soup2.find('table', class_ = "wikitable")
table
```

```python
# Extracting the headers

columns = table.find_all('th',scope = "col")

headers = []

for header in columns:
    headers.append(header.text.strip())
    
print(headers)  # Just to check thw headers
```

```python
# Extracting the States

states = table.find_all('th', scope = 'row')

states_name = []
for state in states:
    states_name.append(state.text.strip())
    
print(states_name)
print(len(states_name))
```

```python
table_rows = table.find_all('tr')
table_rows
```

```python
# Extracting the state_code

state_code = []

for tr in table.find_all('tr'):
    data = []
    for td in tr.find_all('td'): 
        data.append(td.text.strip())
    if data:
        state_code.append(data[0])
        
print("State_code is -",state_code)
```

```python
# Extracting the capital, population and Total area of the state and inserted none in missing values 

data=[]
for i in table.find_all('td'):
    data.append(i.text.strip())
data.insert(18,'None')
data.insert(26,'None')
data.insert(42,'None')
data.insert(74,'None')
data.insert(82,'None')
data.insert(89,'None')
data.insert(106,'None')
data.insert(114,'None')
data.insert(162,'None')
data.insert(186,'None')
data.insert(274,'None')
data.insert(305,'None')
data.insert(329,'None')
data.insert(345,'None')
data.insert(393,'None')
data.insert(282,'None')
data.insert(378,'None')

# Capital occurs after every 8th position so the skip value is 8
capitals=data[1::8]

print("Capitals of states :\n\n",capitals)

Population=data[4::8]

print("\nPopulation of states :\n\n",Population)

Area = data[5::8]
print("\nTotal Area of states :\n\n",Area)
```

```python
# Converting each list into a dictionary

dict1 = {"State": states_name, "State_code": state_code, "Capital":capitals, "Population":Population,"Total_Area":Area}
dict1
```

```python
# To transform the data into a dataframe
import pandas as pd

df = pd.DataFrame(dict1)
df.head()
```

```python
# To convert the dataframe into a csv file

df.to_csv("States.csv", index = False)
```
